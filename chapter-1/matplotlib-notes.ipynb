## Matplotlib
> plotting

```python
# import matplotlib
from matplotlib import pyplot as plt
# import matplotlib.pyplot as plt

# import numpy
import numpy as np
```

### simple plot

#### default settings

```python
X = np.linspace(-np.pi, np.pi, 256) # array containing 256 elements from -pi to pi
C, S = np.cos(X), np.sin(X) # compute cos(x) and sin(x) for all X
```

using only plot(), matplotlib:
- automatically chooses colors
- automatically scales axes
- automatically places ticks

```python
plt.plot(X, C) # graph - (x, cos(x))
plt.plot(X, S) # graph - (x, sin(x))

plt.show()
```

#### subplots
subplot divides the figure into a grid and selects one Axes (matplotlib object) to draw on

```python
# creating a figure (canvas)
plt.figure(figsize = (8, 6), dpi = 80)

# creating one axes inside the figure (1 row, 1 column, position 1)
plt.subplot(1, 1, 1)
```

```python
plt.plot(X, C, color = "blue", linewidth = 1.0, linestyle = "-") # plot cosine with a blue continuous line of width 1 (pixels)
```

```python
plt.plot(X, S, color = "green", linewidth = 1.0, linestyle = "-") # plot sine with a green continuous line of width 1 (pixels)
```

- `linestyle = "-"`  # solid
- `linestyle = "--"` # dashed
- `linestyle = ":"`  # dotted
- `linestyle = "-."` # dash-dot

#### limits
controls the visible region of the axis (zoom level)

```python
# manual limits
plt.figure()

plt.plot(X, C)
plt.plot(X, S)

plt.xlim(-4.0, 4.0)
plt.ylim(-1.0, 1.0)

plt.show()
```

```python
# data driven limits
plt.figure()

plt.plot(X, C)
plt.plot(X, S)

plt.xlim(X.min()*1.1, X.max()*1.1)
plt.ylim(C.min()*1.1, C.max()*1.1)

plt.show()
```

#### ticks
ticks are the small marks; tick labels are the numbers shown

```python
# linear ticks
plt.figure()

plt.plot(X, C)
plt.plot(X, S)

plt.xticks(np.linspace(-4, 4, 9))
plt.yticks(np.linspace(-1, 1, 5))

plt.show()
```

```python
# mathematical ticks
plt.figure()

plt.plot(X, C)
plt.plot(X, S)

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, 1])

plt.show()
```

```python
# ticks with labels
plt.figure()

plt.plot(X, C)
plt.plot(X, S)

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$'])
plt.yticks([-1, 0, 1], [r'$-1$', r'$0$', r'$1$'])

plt.show()
```

- `r''`  raw string (prevents escape character issues)
- `$...$`  renders LaTeX math

#### spines
spines are the axis boundary lines (top, bottom, left, right)

```python
fig, ax = plt.subplots()

ax.plot(X, C)
ax.plot(X, S)

# hide top and right spines
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# move bottom spine to y = 0
ax.spines['bottom'].set_position(('data', 0))
# move left spine to x=0
ax.spines['left'].set_position(('data', 0))
# ('data', 0) means position spine at coordinate 0 in data space

ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

plt.show()
```

#### legends

```python
plt.figure()

plt.plot(X, C, label = "cosine")
plt.plot(X, S, label = "sine")

plt.legend(loc = "upper left")

plt.show()
```

#### annotations

```python
t = 2 * np.pi / 3

plt.figure()
plt.plot(X, C)
plt.plot(X, S)

# vertical helper line
plt.plot([t, t], [0, np.cos(t)], linestyle = "--")

# highlight point
plt.scatter([t], [np.cos(t)])

plt.annotate(r'$cos(\frac{2\pi}{3})=-\frac{1}{2}$', xy = (t, np.cos(t)), xytext = (-90, -50), textcoords = 'offset points', arrowprops = dict(arrowstyle = "->"))

plt.show()
```

annotate():
- xy - point being annotated
- xytext - location of text
- arrowprops - arrow styling
- textcoords = 'offset points'- position text relative to xy in pixels

#### final styling

```python
fig, ax = plt.subplots()

ax.plot(X, C)
ax.plot(X, S)

for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(14)
    label.set_bbox(dict(facecolor = 'white', edgecolor = 'None', alpha = 0.65))
# alpha controls transparency

plt.show()
```

```python
plt.savefig("plot-1.png", dpi = 72) # save figure using 72 dots per inch
```

#### finally

```python
import numpy as np
import matplotlib.pyplot as plt

# create data
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)

# create figure
fig, ax = plt.subplots(figsize = (10, 6), dpi = 80)

# plot cosine and sine
ax.plot(X, C, color = "blue", linewidth = 2.5, linestyle = "-", label = "cosine")
ax.plot(X, S, color = "red",  linewidth = 2.5, linestyle = "-", label = "sine")

# set limits (data driven)
ax.set_xlim(X.min()*1.1, X.max()*1.1)
ax.set_ylim(C.min()*1.1, C.max()*1.1)

# set ticks
ax.set_xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
ax.set_yticks([-1, 0, 1])

# set tick labels
ax.set_xticklabels([r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
ax.set_yticklabels([r'$-1$', r'$0$', r'$+1$'])

# move spines to center
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))

# snsure ticks only on visible spines
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# add legend
ax.legend(loc = 'upper left')

# annotation at t = 2/3
t = 2 * np.pi / 3

# vertical helper lines
ax.plot([t, t], [0, np.cos(t)], color = 'blue', linestyle = "--", linewidth = 1.5)
ax.plot([t, t], [0, np.sin(t)], color = 'red', linestyle = "--", linewidth = 1.5)

# mark points
ax.scatter([t], [np.cos(t)], color = 'blue', s = 50)
ax.scatter([t], [np.sin(t)], color = 'red', s = 50)

# cosine annotation
ax.annotate(r'$cos(\frac{2\pi}{3})=-\frac{1}{2}$', xy = (t, np.cos(t)), xycoords = 'data', xytext = (-90, -50), textcoords = 'offset points', fontsize = 16, arrowprops = dict(arrowstyle = "->", connectionstyle = "arc3,rad=.2"))
# sine annotation
ax.annotate(r'$sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$', xy = (t, np.sin(t)), xycoords = 'data', xytext = (+10, +30), textcoords = 'offset points', fontsize = 16, arrowprops = dict(arrowstyle = "->", connectionstyle = "arc3,rad=.2"))

# improve tick label visibility
for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(14)
    label.set_bbox(dict(facecolor = 'white', edgecolor = 'None', alpha = 0.65))

plt.savefig("plot-1.png", dpi = 72)
plt.show()
```

### other types of plots

#### regular plots
- time series data (stock prices, temperature)
- trends over continuous variables
- signal processing

```python
n = 256
X = np.linspace(-np.pi, np.pi, n)
Y = np.sin(2 * X)

plt.axes([0.025, 0.025, 0.95, 0.95])

plt.plot(X, Y + 1, color = 'blue', alpha = 1.00)
plt.fill_between(X, 1, Y + 1, color = 'blue', alpha = .25)

plt.plot(X, Y - 1, color = 'blue', alpha = 1.00)
plt.fill_between(X, -1, Y - 1, (Y - 1) > -1, color = 'blue', alpha = .25)
plt.fill_between(X, -1, Y - 1, (Y - 1) < -1, color = 'red',  alpha = .25)

plt.xlim(-np.pi, np.pi)
plt.xticks([])
plt.ylim(-2.5, 2.5)
plt.yticks([])

plt.show()
```

#### scatter plots
- relationship between two variables
- correlation analysis
- clustering visualization
- detecting outliers

```python
n = 1024
X = np.random.normal(0, 1, n)
Y = np.random.normal(0, 1, n)
T = np.arctan2(Y, X)

plt.axes([0.025, 0.025, 0.95, 0.95])
plt.scatter(X, Y, s = 75, c = T, alpha = .5)

plt.xlim(-1.5, 1.5)
plt.xticks([])
plt.ylim(-1.5, 1.5)
plt.yticks([])

plt.show()
```

#### bar plots
- comparing categorical data
- frequency comparisons
- survey results
- performance comparison across groups

```python
n = 12
X = np.arange(n)
Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)

plt.axes([0.025, 0.025, 0.95, 0.95])
plt.bar(X, +Y1, facecolor = '#9999ff', edgecolor = 'white')
plt.bar(X, -Y2, facecolor = '#ff9999', edgecolor = 'white')

for x, y in zip(X, Y1):
    plt.text(x + 0.4, y + 0.05, '%.2f' % y, ha = 'center', va = 'bottom')

for x, y in zip(X, Y2):
    plt.text(x + 0.4, -y - 0.05, '%.2f' % y, ha = 'center', va = 'top')

plt.xlim(-.5, n)
plt.xticks([])
plt.ylim(-1.25, 1.25)
plt.yticks([])

plt.show()
```

#### contour plots
- visualizing 3D surfaces in 2D
- optimization problems (loss landscapes)
- temperature/pressure fields
- probability density visualization

```python
def f(x,y):
    return (1 - x / 2 + x**5 + y**3) * np.exp(-x**2 -y**2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
X,Y = np.meshgrid(x, y)

plt.axes([0.025, 0.025, 0.95, 0.95])

plt.contourf(X, Y, f(X, Y), 8, alpha = .75, cmap = plt.cm.hot)
C = plt.contour(X, Y, f(X, Y), 8, colors = 'black', linewidth = .5)
plt.clabel(C, inline = 1, fontsize = 10)

plt.xticks([])
plt.yticks([])
plt.show()
```

#### imshow
- heatmaps
- correlation matrices
- pixel data visualization
- convolutional neural network outputs

```python
def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3 ) * np.exp(-x ** 2 - y ** 2)

n = 10
x = np.linspace(-3, 3, int(3.5 * n))
y = np.linspace(-3, 3, int(3.0 * n))
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

plt.axes([0.025, 0.025, 0.95, 0.95])
plt.imshow(Z, interpolation = 'nearest', cmap = 'bone', origin = 'lower')
plt.colorbar(shrink = .92)

plt.xticks([])
plt.yticks([])
plt.show()
```

#### pie charts
- showing percentage composition
- proportional category distribution
- simple part-to-whole relationships

not ideal for many categories (hard to compare angles)

```python
n = 20
Z = np.ones(n)
Z[-1] *= 2

plt.axes([0.025, 0.025, 0.95, 0.95])

plt.pie(Z, explode = Z*.05, colors = ['%f' % (i / float(n)) for i in range(n)])
plt.axis('equal')
plt.xticks([])
plt.yticks()

plt.show()
```

#### quiver plots
- vector fields
- fluid dynamics
- gradient fields
- electromagnetic field visualization
- direction + magnitude representation

```python
n = 8
X, Y = np.mgrid[0:n, 0:n]
T = np.arctan2(Y - n / 2., X - n/2.)
R = 10 + np.sqrt((Y - n / 2.0) ** 2 + (X - n / 2.0) ** 2)
U, V = R * np.cos(T), R * np.sin(T)

plt.axes([0.025, 0.025, 0.95, 0.95])
plt.quiver(X, Y, U, V, R, alpha = .5)
plt.quiver(X, Y, U, V, edgecolor = 'k', facecolor = 'None', linewidth = .5)

plt.xlim(-1, n)
plt.xticks([])
plt.ylim(-1, n)
plt.yticks([])

plt.show()
```

#### grids
- precise measurement visualization
- scientific plotting
- engineering graphs
- when minor/major tick distinction is required

```python
ax = plt.axes([0.025, 0.025, 0.95, 0.95])

ax.set_xlim(0,4)
ax.set_ylim(0,3)
ax.xaxis.set_major_locator(plt.MultipleLocator(1.0))
ax.xaxis.set_minor_locator(plt.MultipleLocator(0.1))
ax.yaxis.set_major_locator(plt.MultipleLocator(1.0))
ax.yaxis.set_minor_locator(plt.MultipleLocator(0.1))
ax.grid(which = 'major', axis = 'x', linewidth = 0.75, linestyle = '-', color = '0.75')
ax.grid(which = 'minor', axis = 'x', linewidth = 0.25, linestyle = '-', color = '0.75')
ax.grid(which = 'major', axis = 'y', linewidth = 0.75, linestyle = '-', color = '0.75')
ax.grid(which = 'minor', axis = 'y', linewidth = 0.25, linestyle = '-', color = '0.75')
ax.set_xticklabels([])
ax.set_yticklabels([])

plt.show()
```

#### polar axis
- angular data
- directional statistics
- radar-like visualization
- periodic phenomena

```python
ax = plt.axes([0.025, 0.025, 0.95, 0.95], polar = True)

N = 20
theta = np.arange(0.0, 2 * np.pi, 2 * np.pi / N)
radii = 10 * np.random.rand(N)
width = np.pi / 4 * np.random.rand(N)
bars = plt.bar(theta, radii, width=width, bottom = 0.0)

for r,bar in zip(radii, bars):
    bar.set_facecolor(plt.cm.jet(r/10.))
    bar.set_alpha(0.5)

ax.set_xticklabels([])
ax.set_yticklabels([])
plt.show()
```

#### 3d plots
- surface visualization
- terrain models
- loss surface visualization in ML
- scientific simulations

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

ax.plot_surface(X, Y, Z, rstride = 1, cstride = 1, cmap = plt.cm.hot)
ax.contourf(X, Y, Z, zdir = 'z', offset = -2, cmap = plt.cm.hot)
ax.set_zlim(-2, 2)

plt.show()
```
