## Matplotlib

```python
# import matplotlib
from matplotlib import pyplot as plt
# import matplotlib.pyplot as plt

# import numpy
import numpy as np
```

### simple plot

#### default settings

```python
X = np.linspace(-np.pi, np.pi, 256) # array containing 256 elements from -pi to pi
C, S = np.cos(X), np.sin(X) # compute cos(x) and sin(x) for all X
```

using only plot(), matplotlib:
- automatically chooses colors
- automatically scales axes
- automatically places ticks

```python
plt.plot(X, C) # graph - (x, cos(x))
plt.plot(X, S) # graph - (x, sin(x))

plt.show()
```

#### subplots
subplot divides the figure into a grid and selects one Axes (matplotlib object) to draw on

```python
# creating a figure (canvas)
plt.figure(figsize = (8, 6), dpi = 80)

# creating one axes inside the figure (1 row, 1 column, position 1)
plt.subplot(1, 1, 1)
```

```python
plt.plot(X, C, color = "blue", linewidth = 1.0, linestyle = "-") # plot cosine with a blue continuous line of width 1 (pixels)
```

```python
plt.plot(X, S, color = "green", linewidth = 1.0, linestyle = "-") # plot sine with a green continuous line of width 1 (pixels)
```

- `linestyle = "-"`  # solid
- `linestyle = "--"` # dashed
- `linestyle = ":"`  # dotted
- `linestyle = "-."` # dash-dot

#### limits
controls the visible region of the axis (zoom level)

```python
# manual limits
plt.figure()

plt.plot(X, C)
plt.plot(X, S)

plt.xlim(-4.0, 4.0)
plt.ylim(-1.0, 1.0)

plt.show()
```

```python
# data driven limits
plt.figure()

plt.plot(X, C)
plt.plot(X, S)

plt.xlim(X.min()*1.1, X.max()*1.1)
plt.ylim(C.min()*1.1, C.max()*1.1)

plt.show()
```

#### ticks
ticks are the small marks; tick labels are the numbers shown

```python
# linear ticks
plt.figure()

plt.plot(X, C)
plt.plot(X, S)

plt.xticks(np.linspace(-4, 4, 9))
plt.yticks(np.linspace(-1, 1, 5))

plt.show()
```

```python
# mathematical ticks
plt.figure()

plt.plot(X, C)
plt.plot(X, S)

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, 1])

plt.show()
```

```python
# ticks with labels
plt.figure()

plt.plot(X, C)
plt.plot(X, S)

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$'])
plt.yticks([-1, 0, 1], [r'$-1$', r'$0$', r'$1$'])

plt.show()
```

- `r''`  raw string (prevents escape character issues)
- `$...$`  renders LaTeX math

#### spines
spines are the axis boundary lines (top, bottom, left, right)

```python
fig, ax = plt.subplots()

ax.plot(X, C)
ax.plot(X, S)

# hide top and right spines
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# move bottom spine to y = 0
ax.spines['bottom'].set_position(('data', 0))
# move left spine to x=0
ax.spines['left'].set_position(('data', 0))
# ('data', 0) means position spine at coordinate 0 in data space

ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

plt.show()
```

#### legends

```python
plt.figure()

plt.plot(X, C, label = "cosine")
plt.plot(X, S, label = "sine")

plt.legend(loc = "upper left")

plt.show()
```

#### annotations

```python
t = 2 * np.pi / 3

plt.figure()
plt.plot(X, C)
plt.plot(X, S)

# vertical helper line
plt.plot([t, t], [0, np.cos(t)], linestyle = "--")

# highlight point
plt.scatter([t], [np.cos(t)])

plt.annotate(r'$cos(\frac{2\pi}{3})=-\frac{1}{2}$', xy = (t, np.cos(t)), xytext = (-90, -50), textcoords = 'offset points', arrowprops = dict(arrowstyle = "->"))

plt.show()
```

annotate():
- xy - point being annotated
- xytext - location of text
- arrowprops - arrow styling
- textcoords = 'offset points'- position text relative to xy in pixels

#### final styling

```python
fig, ax = plt.subplots()

ax.plot(X, C)
ax.plot(X, S)

for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(14)
    label.set_bbox(dict(facecolor = 'white', edgecolor = 'None', alpha = 0.65))
# alpha controls transparency

plt.show()
```

```python
plt.savefig("plot-1.png", dpi = 72) # save figure using 72 dots per inch
```

#### finally

```python
import numpy as np
import matplotlib.pyplot as plt

# create data
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)

# create figure
fig, ax = plt.subplots(figsize = (10, 6), dpi = 80)

# plot cosine and sine
ax.plot(X, C, color = "blue", linewidth = 2.5, linestyle = "-", label = "cosine")
ax.plot(X, S, color = "red",  linewidth = 2.5, linestyle = "-", label = "sine")

# set limits (data driven)
ax.set_xlim(X.min()*1.1, X.max()*1.1)
ax.set_ylim(C.min()*1.1, C.max()*1.1)

# set ticks
ax.set_xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
ax.set_yticks([-1, 0, 1])

# set tick labels
ax.set_xticklabels([r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
ax.set_yticklabels([r'$-1$', r'$0$', r'$+1$'])

# move spines to center
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))

# snsure ticks only on visible spines
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# add legend
ax.legend(loc = 'upper left')

# annotation at t = 2/3
t = 2 * np.pi / 3

# vertical helper lines
ax.plot([t, t], [0, np.cos(t)], color = 'blue', linestyle = "--", linewidth = 1.5)
ax.plot([t, t], [0, np.sin(t)], color = 'red', linestyle = "--", linewidth = 1.5)

# mark points
ax.scatter([t], [np.cos(t)], color = 'blue', s = 50)
ax.scatter([t], [np.sin(t)], color = 'red', s = 50)

# cosine annotation
ax.annotate(r'$cos(\frac{2\pi}{3})=-\frac{1}{2}$', xy = (t, np.cos(t)), xycoords = 'data', xytext = (-90, -50), textcoords = 'offset points', fontsize = 16, arrowprops = dict(arrowstyle = "->", connectionstyle = "arc3,rad=.2"))
# sine annotation
ax.annotate(r'$sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$', xy = (t, np.sin(t)), xycoords = 'data', xytext = (+10, +30), textcoords = 'offset points', fontsize = 16, arrowprops = dict(arrowstyle = "->", connectionstyle = "arc3,rad=.2"))

# improve tick label visibility
for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(14)
    label.set_bbox(dict(facecolor = 'white', edgecolor = 'None', alpha = 0.65))

plt.savefig("plot-1.png", dpi = 72)
plt.show()
```
